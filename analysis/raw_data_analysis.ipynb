```python
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
# user histograms on raw_data

with open('../raw_data/ratings.dat', newline='') as file:
    rows = csv.reader(file, delimiter='^', quotechar='|')
    dic = {}
    for idx, row in enumerate(rows):
        user = int(row[0])
        dic[user] = dic.get(user, 0) + 1
    
    l = list(dic.values())
    l.sort(reverse=True)
    plt.plot(np.array(l))
    plt.ylabel('users ratings counts')
    plt.show()
    
    n_users = len(l)
    percent_users = [0.0001, 0.001, 0.01, 0.1, 0.5]
    for percent_user in percent_users:
        print("{}% users rate less than {} movies".format((1-percent_user) * 100, l[int(percent_user*n_users)]))
```

```python
#item histograms on raw data

with open('../raw_data/ratings.dat', newline='') as file:
    rows = csv.reader(file, delimiter='^', quotechar='|')
    dic = {}
    for idx, row in enumerate(rows):
        movie = int(row[1])
        dic[movie] = dic.get(movie, 0) + 1
    
    l = list(dic.values())
    l.sort(reverse=True)
    plt.plot(np.array(l))
    plt.ylabel('movies ratings counts')
    plt.show()
    
    n_movies = len(l)
    percent_movies = [0.0001, 0.001, 0.01, 0.1, 0.5]
    for percent_movie in percent_movies:
        print("{}% movies was rated less than {} times".format((1-percent_movie) * 100, l[int(percent_movie*n_movies)]))
```
