```python
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
# user histograms on train_data.dat and test_data.dat

with open('../data/train_data.dat', newline='') as file:
    rows = csv.reader(file, delimiter=':', quotechar='|')
    dic = {}
    for idx, row in enumerate(rows):
        user = int(row[0])
        dic[user] = dic.get(user, 0) + 1
    
    l = list(dic.values())
    l.sort(reverse=True)
    plt.plot(np.array(l))
    plt.ylabel('users ratings counts of train_data.dat')
    plt.show()
    
    n_users = len(l)
    percent_users = [0.0001, 0.001, 0.01, 0.1, 0.5]
    for percent_user in percent_users:
        print("{}% users rate less than {} movies".format((1-percent_user) * 100, l[int(percent_user*n_users)]))

with open('../data/test_data.dat', newline='') as file:
    rows = csv.reader(file, delimiter=':', quotechar='|')
    dic = {}
    for idx, row in enumerate(rows):
        user = int(row[0])
        dic[user] = dic.get(user, 0) + 1
    
    l = list(dic.values())
    l.sort(reverse=True)
    plt.plot(np.array(l))
    plt.ylabel('users ratings counts of test_data.dat')
    plt.show()
#TODO: calculate more stats of users
```
